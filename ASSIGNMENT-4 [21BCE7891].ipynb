```python
# precision
# of all positive predictions how many are really positive

# precision = TP/(TP+FP)
Precision = (TP)/(TP+FP)
Precision
```

```python
# Recall
# of all real positive cases how many are predicted positive

# Recall = TP/(FN+TP)
Recall = (TP)/(FN+TP)
Recall
```

```python
# F1 score

# 2*precision*Recall/(Precision+Recall)
F1_Score = (2*Precision*Recall)/(Precision+Recall)
F1_Score
```
