# DATA VISUALISATION

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
dataset = sns.load_dataset('car_crashes')
dataset
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

### LINE GRAPH

```python
plt.figure(figsize = (20, 6))
sns.lineplot(x = "abbrev", y = "total", data = dataset, marker = "*",markersize = 15, markerfacecolor="red")
```

Inference: The percentage of drivers involved in car crashes are found to be the lowest in the state abbreviated with "DC" and the highest in the states abbreviated with "ND", "SC" and "WV".

### HISTOGRAM

```python
plt.hist(dataset["alcohol"])
```

Inference: The percentage of drivers involved in car crashes under the influence of alcohol is distributed as shown in the figure. There is no frequency at 8.349 and 9.1935

```python
plt.hist(dataset["speeding"])
```

Inference: The percentage of drivers involved in car crashes while speeding is distributed as shown in the figure.

### SCATTER PLOT

```python
sns.scatterplot(x="alcohol", y="total", data=dataset)
plt.title("Alcohol Consumption vs Total Crashes")
plt.xlabel("Alcohol Consumption")
plt.ylabel("Total Crashes")
plt.show()
```

Inference: Higher the alcohol consumption, higher the percentage of drivers involved in car crashes

```python
sns.scatterplot(x="speeding", y="total", data=dataset)
plt.title("Speed vs Total Crashes")
plt.xlabel("Speeding")
plt.ylabel("Total Crashes")
plt.show()
```

Inference: Higher the speeding level, higher the percentage of drivers involved in car crashes

```python
sns.scatterplot(x="not_distracted", y="total", data=dataset)
plt.title("Distractionless vs Total Crashes")
plt.xlabel("Distractionlessness")
plt.ylabel("Total Crashes")
plt.show()
```

Inference: Lesser the number of distractions, more is the percentage of drivers involved in car crashes. 
           This can be due to driver sleeping/overspeeding in the car by the virtue of lesser distractions.

```python
sns.scatterplot(x="no_previous", y="total", data=dataset)
plt.title("No previous crashes vs Total Crashes")
plt.xlabel("Having no previous crashes")
plt.ylabel("Total Crashes")
plt.show()
```

Inference: Percentage of drivers having no previous car crashes contribute more to the percentage of drivers involved in car crashes. This can be due to lack of experience and safety.

### BAR PLOT

```python
datasetdesc = dataset.sort_values("total", ascending=False)
f, a = plt.subplots(figsize=(20, 6))
sns.barplot(x="abbrev", y="total", data=datasetdesc, label="Total crashes", color="pink")
sns.barplot(x="abbrev", y="alcohol", data=datasetdesc, label="Alcohol induced crashes", color="hotpink")
percentages = (datasetdesc["alcohol"] / datasetdesc["total"]) * 100
for i, p in enumerate(a.patches[len(dataset):]):
    a.text(p.get_x() + p.get_width() / 2, p.get_height() + 0.5, f"{percentages.iloc[i]:.1f}%", ha='center', fontsize=12, rotation = 90, color='hotpink')
for i, p in enumerate(a.patches[len(dataset):]):
    a.text(p.get_x() + p.get_width() / 2, p.get_height() + 0.5, f"{percentages.iloc[i]:.1f}%", ha='center', fontsize=12, rotation = 90, color='pink')
a.legend(ncol=2, loc="upper right", frameon=True)
a.set(ylim=(0, 26), ylabel="Total Crashes",xlabel="State name abbreviations")
```

Inference: The state with the abbreviation "MT" has 44%(highest percentage of total) of drivers involved in car crashes due to alcohol consumption as opposed to the state with the abbreviation "UT" that has 16%(lowest percentage of total) of drivers involved car crashes due to alcohol consumption. The total percentage of drivers involved in car crashes are lowest in the state with the abbreviation "DC" and highest in the state abbreviated with "SC" and "ND".

### PIE CHART

```python
abbreviations = dataset['abbrev']
total_crashes = dataset['total']
num_colors = 51
clrs = sns.color_palette("husl", num_colors)
plt.figure(figsize=(15, 17))
plt.pie(total_crashes, labels=abbreviations, colors=clrs, autopct='%1.1f%%')
plt.title("Car Crashes by State Abbreviation")
plt.axis('equal')
plt.show()
```

Inference: The states with abbreviations "WV", "SC" and "ND"  each contribute to 3% of the drivers involved in car crashes which is higher than other states

### HEATMAP

```python
cor = dataset.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(cor, annot = True, cmap = "YlGnBu")
```

Inference: The total percentage of drivers involved in car crashes are highly correlated with having no previous crashes, not being distracted, alcohol and moderately with speeding. It is negatively correlated insurance premium and insurance losses.

### RELATIONAL PLOT

```python
sns.relplot(x = "total", y = "alcohol", data = dataset, hue = "abbrev")
```

Inference: The above plot shows the relation between the total percentage of drivers involved in a car crash and the percentage of drivers involved in a car crash and under the influence of alcohol. Lowest being the state abbreviated as "DC" and highest being the state abbreviated as "ND".

```python
sns.relplot(x = "total", y = "speeding", data = dataset, hue = "abbrev")
```

Inference: The above plot shows the relation between the total percentage of drivers involved in a car crash and the percentage of drivers involved in a car crash and are speeding.

### JOINT PLOT

```python
sns.jointplot(x = "alcohol", y = "speeding", data = dataset)
```

Inference: The above plot shows the positive correlation between the total percentage of drivers involved in a car crash and under the influence of alcohol and the percentage of drivers involved in a car crash and overspeeding. The histograms of the axes are also plotted.
